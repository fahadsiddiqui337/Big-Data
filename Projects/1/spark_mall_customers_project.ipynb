# Mall Customers Analysis using Apache Spark

Step-by-step mini project with explanations.

## Step 1: Start Spark Session
Roman Urdu: Yahan hum Spark ka engine start karte hain.

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Mall Customer Analysis').getOrCreate()
```

## Step 2: Load Dataset
Roman Urdu: Dataset ko Spark me load kar rahe hain.

```python
df = spark.read.csv('Mall_Customers(1).csv', header=True, inferSchema=True)
df.show()
```

## Step 3: Check Schema
Roman Urdu: Data ka structure dekhte hain.

```python
df.printSchema()
```

## Step 4: Basic Statistics
Roman Urdu: Average, min, max values dekhte hain.

```python
df.describe().show()
```

## Step 5: Select Columns
Roman Urdu: Sirf specific columns dekhte hain.

```python
df.select('Age', 'Annual Income (k$)').show()
```

## Step 6: Filter Data
Roman Urdu: Condition apply karte hain.

```python
df.filter(df['Age'] > 30).show()
```

## Step 7: Group By Analysis
Roman Urdu: Gender ke basis par analysis.

```python
df.groupBy('Genre').avg('Spending Score (1-100)').show()
```

## Step 8: Sorting Data
Roman Urdu: Highest income walay top par.

```python
df.orderBy('Annual Income (k$)', ascending=False).show()
```

## Step 9: New Column
Roman Urdu: Income me 10% increase.

```python
df = df.withColumn('Bonus Income', df['Annual Income (k$)'] * 1.1)
df.show()
```

## Step 10: Count
Roman Urdu: Total customers.

```python
df.count()
```

## Step 11: High Value Customers
Roman Urdu: Rich aur high spending customers.

```python
df.filter((df['Annual Income (k$)'] > 70) & (df['Spending Score (1-100)'] > 70)).show()
```
